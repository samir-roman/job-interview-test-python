## Importar librería "pandas"

```python
import pandas as pd
```

## Leer el archivo ventas_compras.xlsx 

```python
df = pd.read_excel('ventas_compras.xlsx', sheet_name = [0,1])
```

## Visualizar los datos de Ventas y Compras

```python
df_ventas = df[0] 
df_ventas.head()
```

```python
df_compras = df[1]
df_compras.head()
```

## Total de ventas por cliente 

```python
df_ventas['total_ventas_cliente'] = df_ventas['Producto']

df_total_ventas = df_ventas[['Cliente', 'total_ventas_cliente']].groupby(by=['Cliente']).count()
df_total_ventas
```

## Total de compras por proveedor

```python
df_compras['total_compras_proveedor'] = df_compras['Producto']

df_total_compras = df_compras[['Proveedor', 'total_compras_proveedor']].groupby(by=['Proveedor']).count()
df_total_compras
```

## Exportar a Excel las 2 hojas

```python
with pd.ExcelWriter('total_ventas_compras.xlsx') as writer:
    df_total_ventas.to_excel(writer,\
                            sheet_name="Total de Ventas por Cliente")
    df_total_compras.to_excel(writer,\
                            sheet_name="Total de Compras por Proveedor")
```
